```python
from PIL import Image
import os

def resize_icons(source_path, output_folder):
    # 确保输出目录存在
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # 目标尺寸
    sizes = [192, 512]
    
    try:
        with Image.open(source_path) as img:
            # 确保是正方形
            if img.size[0] != img.size[1]:
                print("警告：原图不是正方形，可能会产生拉伸。")
            
            for size in sizes:
                # 使用高质量重采样 (LANCZOS)
                resized_img = img.resize((size, size), Image.Resampling.LANCZOS)
                output_name = f"icon-{size}.png"
                resized_img.save(os.path.join(output_folder, output_name))
                print(f"成功生成: {output_name} ({size}x{size})")
                
    except Exception as e:
        print(f"发生错误: {e}")

# 执行转换
# 假设你的原图叫 icon_source.png，存放在当前目录下
# 输出到 public 文件夹
resize_icons("../../Downloads/icon-1024.png", "../../Downloads")
```
